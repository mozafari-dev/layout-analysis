```python
for index, rows in df.iterrows():
    X = float(rows['x'])
    Y = float(rows['y'])
    _X = (210 * X)/1653
    _Y = (297 * Y)/2139
    df.loc[index, 'x'] = round(_X, 2)
    df.loc[index, 'y'] = round(_Y, 2)
```

```python
# finding distance and backward reading in the data
featured_list = list()
x1 = x2 = y1 = y2 = dist = regression = speed = 0

for index, rows in df.iterrows():
    if index == 0:
        x1 = rows['x']
        y2 = rows['y']
        featured_list.append([rows['time'], rows['fix_i'], round(rows['x'], 2), round(rows['y'], 2), rows['fix_dur'], rows['doc'], 
                              rows['subject'], rows['cond'], rows['section'], rows['roi'], dist, regression])
    else:
        x2 = rows['x']
        y2 = rows['y']
        dist = round(math.sqrt(((x2-x1)**2 + (y2-y1)**2)), 2)
        featured_list.append([rows['time'], rows['fix_i'], round(rows['x'], 2), round(rows['y'], 2), rows['fix_dur'], rows['doc'], 
                              rows['subject'], rows['cond'], rows['section'], rows['roi'], dist, regression])

        if (x2 - x1) < 0:
            regression = 1
        else:
            regression = 0
        
        x1 = rows['x']
        y1 = rows['y']
        
        
```

```python
# Calculation of Regression, Sweep return
fdf = pd.DataFrame(featured_list)
fdf.rename(columns={0: 'time', 1: 'fix_index', 2: 'x', 3: 'y', 4: 'fix_dur', 5: 'doc', 
                    6: 'subject', 7: 'cond', 8: 'section', 9: 'roi', 10: 'dist', 11: 'backward'}, inplace=True)

fdf.insert(loc=12, column='regression', value=0)
fdf.insert(loc=13, column='sweep', value=0)
fdf.head()
```

```python
for index, rows in fdf.iterrows():
    if index > 0:
        _time = df.loc[index, 'time'] - df.loc[(index - 1), 'time'] 
    if  rows['backward'] == 1:
        if rows['dist'] >= 20:
            fdf.loc[index, 'sweep'] = 1
        else:
            fdf.loc[index, 'regression'] = 1

        
#Save in separate CSV file        
fdf.to_csv('all_fixations_with_parameters_2.csv', index = False)  
```

```python
fdf = pd.read_csv('all_fixations_with_parameters_2.csv')
```

```python
# Forward Reading 
fr_flag = fr_index = fr_distance = fr_duration = fr_speed = fr_time = t1 = tn = 0
fr_list = list()

for index, rows in fdf.iterrows():
   
    if rows['backward'] == 0 and fr_flag == 0:
        t1 = rows['time']
        fr_speed += 0
        fr_flag = 1
        fr_index += 1
        fr_distance += rows['dist']
        fr_duration += rows['fix_dur']

    elif rows['backward'] == 0 and fr_flag == 1:
        tn = rows['time']
        fr_distance += rows['dist']
        fr_duration += rows['fix_dur']

    elif rows['backward'] != 0 and fr_flag == 1:
        fr_time = tn - t1
        if fr_time == 0:
            fr_time = 0.5
        fr_speed = round(fr_distance / fr_time, 2)
        fr_list.append([fr_index, fr_distance, fr_time, fr_duration,fr_speed,
                        rows['cond'], rows['subject'], rows['section']])
        fr_flag = fr_distance = fr_duration = 0

    else:
        continue

fr_df = pd.DataFrame(fr_list)
fr_df.rename(columns={0: 'fr_index', 1: 'fr_dist', 2: 'fr_time', 3: 'fr_duration', 4: 'fr_speed', 
                      5: 'cond', 6: 'subject', 7: 'section'}, inplace=True)

# Save in separate CSV file        
```
