```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_data():
    np.random.seed(24)
    N = 1000
    shift_value = -6.

    x2 = np.random.randn(N) + shift_value
    y2 = np.random.randn(N) + shift_value
    return x2, y2

def cart2polar(x,y):
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y/2,x)

    return theta, r
```

```python
x2, y2 = generate_data()
theta2, r2 = cart2polar(x2,y2)

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection="polar")
ax2.scatter(theta2, r2, color='hotpink')
ax2.scatter(np.abs(theta2), r2, color='hotpink')

H, xedges, yedges = np.histogram2d(np.hstack([x2,x2]),np.hstack([y2,-y2]))

xedges, yedges = np.meshgrid(xedges[:-1], yedges[:-1])
theta_edges, r_edges = cart2polar(xedges, yedges)
ax2.contour(theta_edges, r_edges, H)

plt.show()
```
