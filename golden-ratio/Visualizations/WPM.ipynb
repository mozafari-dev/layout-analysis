```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
sns.set(style="white")
```

```python
df = pd.read_csv('../participants.csv', sep=',')
participants = df.participant.unique()
```

```python
sf = pd.DataFrame(columns=['participant', 'golden_ratio', 'compact', 'total'])

for i, p in enumerate(participants):
    f = df[df.participant==p]
    p = 'P0'+str(p) if len(str(p)) == 1 else 'P'+str(p)
    sf.loc[i] = [p, int(f[f.condition==1].wpm), int(f[f.condition==2].wpm), int(f[f.condition==1].wpm) + int(f[f.condition==2].wpm)]
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(16, 4))

# Plot the total crashes
sns.set_color_codes("muted")
sns.barplot(x="golden_ratio", y="participant", data=sf,
            label="Golden Ratio", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("pastel")
sns.barplot(x="compact", y="participant", data=sf,
            label="Compact", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="upper right", frameon=True)
ax.set(xlim=(0, 560), ylabel="Participants",
       xlabel="WPM")
sns.despine(left=True, bottom=True)
```
